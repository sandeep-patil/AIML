## 🧠 What is Polynomial Regression?

### In simple words:

> It’s **Linear Regression** with **curved lines**!

It lets us fit data that **isn’t straight** — like a curve, bend, or U-shape — by **adding powers of x** (like $x^2, x^3$) to our model.

---

## 📈 Example

Let’s say you plot:

| Speed (x) | Fuel Efficiency (y) |
| --------- | ------------------- |
| 10 km/h   | 5 km/l              |
| 30 km/h   | 8 km/l              |
| 50 km/h   | 9.5 km/l            |
| 70 km/h   | 8 km/l              |
| 90 km/h   | 6 km/l              |

🟥 This does **not look like a straight line**
🟢 It looks like a **curve** — fuel efficiency increases, then decreases.

### 👉 Linear regression **won’t work well**

But if we add $x^2$ (square of speed), the model can **curve to fit the data** better!

---

## 🧮 Equation of Polynomial Regression

For degree 2:

$$
\hat{y} = w_0 + w_1 x + w_2 x^2
$$

For degree 3:

$$
\hat{y} = w_0 + w_1 x + w_2 x^2 + w_3 x^3
$$

Still called "linear" because it's **linear in weights**, even though **nonlinear in x**.

---

## 📦 How It Works:

1. You **create extra features**: $x^2, x^3, \dots$
2. Then use **ordinary linear regression** on this expanded set
3. Output is a **curve**, not a line

---

## 🚘 Automotive Examples:

| Use Case                                     | Why Polynomial Regression?                |
| -------------------------------------------- | ----------------------------------------- |
| Predicting **brake distance** based on speed | Distance ∝ speed²                         |
| Modeling **engine heat** vs RPM              | May rise and flatten at high RPM          |
| SOC vs Voltage curve                         | Batteries have nonlinear discharge curves |
| Torque vs Throttle                           | Curve changes with RPM load               |

---

## 🎓 Analogy: Bending a Ruler

* Linear Regression: Ruler stays straight
* Polynomial Regression: You **bend** the ruler to follow the curve of the data
