## **nonlocal and global**

````python
x = 1

def f1():
    x = 10
    def f2():
        nonlocal x  # Refers to x from f1(), not the global one
        x += 1
        print(x)  # #1

    f2()
    print(x)      # #2

f1()
print(x)          # #3
````

````python
x = 1

def f1():
    x = 10
    def f2():
        global x  # Refers to x from f1(), not the global one
        x += 1
        print(x)  # #1

    f2()
    print(x)      # #2

f1()
print(x)          # #3
````

#**Lambda Funions - Anonymous function**

###  What is a Lambda Function?

* A **lambda** is a **small anonymous function**.
* It has **no name**, defined using the `lambda` keyword.
* Syntax:

  ```python
  lambda arguments: expression
  ```

---

###  Example:

```python
f = lambda x: x + 10
print(f(5))   # Output: 15
```

Same as:

```python
def f(x):
    return x + 10
```

---

###  Multiple Arguments:

```python
add = lambda a, b: a + b
print(add(2, 3))  # Output: 5
```

---

###  Where it's used:

* Quick use in `map()`, `filter()`, `sorted()`, etc.

Example:

```python
nums = [2, 4, 1]
sorted(nums, key=lambda x: -x)  # Sort descending
```

---

###  Remember:

* One expression only (no `return`, no multiple lines)
* Used for **short, throwaway** functions

##  **1. `map()`**

###  Purpose:

Applies a function to **every element** of an iterable (like list, tuple).

###  Syntax:

```python
map(function, iterable)
```

###  Example:

```python
nums = [1, 2, 3]
squared = list(map(lambda x: x**2, nums))
# Output: [1, 4, 9]
```

---

##  **2. `filter()`**

###  Purpose:

Filters elements from an iterable **based on a condition** (returns only elements for which the function is `True`).

###  Syntax:

```python
filter(function, iterable)
```

###  Example:

```python
nums = [1, 2, 3, 4]
even = list(filter(lambda x: x % 2 == 0, nums))
# Output: [2, 4]
```

---

##  **3. `reduce()`**

###  Purpose:

Applies a function **cumulatively** to the items in a sequence (i.e., reduces the iterable to a single value).

###  Syntax:

```python
from functools import reduce
reduce(function, iterable)
```

###  Example:

```python
from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
# Output: 24 (1*2*3*4)
```

---

##  Summary Table:

| Function   | Purpose                               | Returns                    |
| ---------- | ------------------------------------- | -------------------------- |
| `map()`    | Apply function to all items           | New iterable (transformed) |
| `filter()` | Select items that meet condition      | New iterable (filtered)    |
| `reduce()` | Collapse to single result using logic | Single value (aggregated)  |

### **Singleton nature of modules**

**Modules are singleton**:

Once loaded, theyre cached in `sys.modules`.
* **Re-import doesnt re-run the code**.
* **Use `importlib.reload()`** to reflect changes after editing a module.


# For Example
`import mod`


The module mod.py is imported for the first time. Python executes the module and loads its variables/functions.

Suppose mod.py contains: a = 100

mod.a  100

Accessing variable a from mod.

import mod again

No re-execution happens. Python reuses the already loaded mod from sys.modules.

`importlib.reload(mod)`

This forces Python to re-execute the code in mod.py.

Lets assume mod.py was modified and now contains: a = 300

mod.a  300

Shows the new value after reloading.
